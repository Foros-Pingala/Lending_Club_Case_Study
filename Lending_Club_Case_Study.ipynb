```python
def plt_frequency(x_var,data_df,x_label,title,hue_var=None,rot_flag=0,fig_size_x=15,fig_size_y=6):
    ncount = len(loan)
    dims = (fig_size_x,fig_size_y)
    plt.figure(figsize=dims)
    ax = sns.countplot(x=x_var, data=data_df,hue=hue_var,palette="Set2")
    plt.title(title)
    plt.xlabel(x_label)

    # Make twin axis
    ax2=ax.twinx()

    # Switch so count axis is on right, frequency on left
    ax2.yaxis.tick_left()
    ax.yaxis.tick_right()

    # Also switch the labels over
    ax.yaxis.set_label_position('right')
    ax2.yaxis.set_label_position('left')
    if(rot_flag==1):
        ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=12)
        plt.tight_layout()
    ax2.set_ylabel('Frequency [%]')

    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text

    # Need to turn the grid on ax2 off, otherwise the gridlines end up on top of the bars
    ax2.grid(None)
```

### Analysing data from the derived columns
#### dti(debt-to-income ratio): Higher the DTI lower will be chances of loan approval
#### l_t_ai(Loan to annual income): l_t_ai id directly proportion to default rate

```python
def dti_range_calculator(dti):
    if dti <= 10.00:
        return 'low'
    elif (dti >10.00 and dti <= 20.00):
        return 'medium'
    else:
        return 'high'
        
def loan_to_annual_income_range_calculator(l_t_ai):
    if l_t_ai <= 0.10:
        return 'low'
    elif (l_t_ai >0.10 and l_t_ai <= 0.20):
        return 'medium'
    else:
        return 'high'

def experience_level(exp):
    if exp <= 3:
        return 'Junior'
    elif (exp >3 and exp <= 7):
        return 'Senior'
    else:
        return 'Specialist'
    
# category_list = ['debt_consolidation','credit_card','other','home_improvement','major_purchase','small_business']
# loan = loan.query('purpose == @category_list')
loan = loan[loan['purpose'].isin(['debt_consolidation','credit_card','other','home_improvement','major_purchase','small_business'])]
loan['dti_range'] = loan['dti'].apply(lambda x : dti_range_calculator(x) )
loan['l_t_ai_range'] = loan['l_t_ai'].apply(lambda x : loan_to_annual_income_range_calculator(x) )
loan['exp_level'] = loan['emp_length'].apply(lambda x : experience_level(int(x)))
loan['int_rate_float'] = loan['int_rate'].apply(lambda x : float(x.replace('%','')))
# loan['ln_verified'] = loan['verification_status'].groupby()
#loans['int_rate']
```

```python
plt_frequency("loan_status",loan,'Loan status','Average distribution of loans across category and experience level','exp_level')
plt_frequency("loan_status",loan,'Loan Status','Average distribution of loans across category and dti','dti_range')
plt_frequency("loan_status",loan,'Loan Status','Average distribution of loans across category and loan to annual income range','l_t_ai_range')
plt_frequency("loan_status",loan,'Loan Status','Average distribution of loans across category and Grades','grade')
plt_frequency("loan_status",loan[loan['verification_status']=='Verified'],'Loan Status','Average distribution of loans verified across category and Grades','grade')
plt_frequency("loan_status",loan[loan['verification_status']=='Not Verified'],'Loan Status','Average distribution of loans Not Verified across category and Grades','grade')
plt.show()
print("1) From the above plot we can see that people having decent experience have high chances of default ")
print("2) People Having higher dti have lower chances of default")
print("3) People having higher annual income range have higher chances of default")
print("4) Grade B,C and D have higher probability of defaulting")
print("5) Verified loans distributed across the grade B have higher chances of default")
print("6) Not Verified loans distributed across the grade B & C have higher chances of default")
```
